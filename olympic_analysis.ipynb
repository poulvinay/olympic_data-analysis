What are the most Successful Countries of all times?

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
olympic.Country.value_counts()
```

```python
top_10 = olympic.Country.value_counts().head(10)

top_10
```

```python
top_10.plot(kind="bar", fontsize=15, figsize=(12, 8))
plt.title("Top 10 Countries by Medal", fontsize=15)
plt.ylabel("Medal", fontsize=14)
plt.show()
```

```python
olympic_10 = olympic[olympic.Country.isin(top_10.index)]

olympic_10
```

```python
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5, palette="dark")
sns.countplot(data=olympic_10, x="Country", order=top_10.index)
plt.title("Top 10 Country by Medal", fontsize=20)
plt.ylabel("Medal", fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5, palette="dark")
sns.countplot(data=olympic_10, x="Country", hue="Edition", order=top_10.index)
plt.title("Top 10 Country by Medal", fontsize=20)
plt.ylabel("Medal", fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5, palette="dark")
sns.countplot(data=olympic_10, x="Edition", hue="Country", hue_order=top_10.index)
plt.title("Top 10 Country by Medal", fontsize=20)
plt.ylabel("Medal", fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5, palette="dark")
sns.countplot(data=olympic_10, x="Country", hue="Medal", order=top_10.index)
plt.title("Top 10 Country by Medal", fontsize=20)
plt.ylabel("Medal", fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5, palette="dark")
sns.countplot(data=olympic_10, x="Country", hue="Medal", order=top_10.index, hue_order=["Gold", "Silver", "Bronze"], palette=["Gold", "Silver", "Brown"])
plt.title("Top 10 Country by Medal", fontsize=20)
plt.ylabel("Medal", fontsize=15)
plt.show()
```
